```python
# 리스트로 만들고 정렬 1:1로 맞아떨어진다...?
words = ['eat', 'tea', 'tan', 'ate', 'nat', 'bat', 'bee', 'eb', 'ebe']
```

```python
set_list = []
# 애너그램 리스트 만들기
for word in words:
    # 정렬 후 튜플로 받아 리스트에 추가
    set_list.append(tuple(sorted(word)))

# set으로 중복 제거 후 애너그램 리스트 생성
vs_list = list(set(set_list))

res_dict = dict()
for word in words:
    for key in vs_list:
        if tuple(sorted(word)) == key:
            res_dict.setdefault(key, [])
            res_dict[key].append(word)
            
list(res_dict.values())
```

```python
# 원석님
def anagram():
    set_a = set()
    for i in range(len(words)):
        tmp = list(words[i])
        tmp.sort()
        set_a.add(''.join(tmp))
    return list(set_a)

words = ['eat', 'tea', 'tan', 'ate', 'nat', 'bat']
print(anagram())
```

```python
# 강사님
def anagram(words):
    anagrams = dict()
    result = []
    # 애너그램에 해당되는 값들을 서로 비교해서
    # 같은 그룹이라면 해당 그룹에 리스트로 포함
    for word in words:
        # 문자열 -> 리스트(가변)화 => 장랼 -> 애너그램 그룹
        # 불변형 변환(tuple, str)
        key = ''.join(sorted(list(word)))
        anagrams.setdefault(key, [])
        anagrams[key].append(word)

    return [value for value in anagrams.values()]

words = ['eat', 'tea', 'tan', 'ate', 'nat', 'bat']
print(anagram(words))
```
